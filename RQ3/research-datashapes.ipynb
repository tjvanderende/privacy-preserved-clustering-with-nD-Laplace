## Research breast cancer dataset


```python
import pandas as pd
from Helpers import helpers
import numpy as np
import matplotlib.pyplot as plt

breast_cancer_df = helpers.load_dataset('../data/breast-cancer.csv')
```

```python
breast_cancer_df.head()
```

```python
twod_columns = ['area_worst', 'concavity_worst', 'class']
three_columns = ['area_worst', 'concavity_worst', 'radius_worst', 'class']
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(breast_cancer_df['class'])
breast_cancer_df['class'] = le.transform(breast_cancer_df['class'])
```

```python
threed_df = breast_cancer_df[three_columns]
twod_df = breast_cancer_df[twod_columns]
```

```python
threed_df.head()
```

```python
twod_df.head()
```

```python
twod_df.plot(kind='scatter', x='area_worst', y='concavity_worst', c='class', colormap='viridis')
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.scatter3D(threed_df.area_worst, threed_df.concavity_worst, threed_df.radius_worst, c=threed_df['class'])
```
