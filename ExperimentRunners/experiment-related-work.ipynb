```python
import pandas as pd
import numpy as np
related_work_df = pd.read_csv('related-work.csv')
related_work_df.head()
```

```python
related_work_df = related_work_df.convert_dtypes()
related_work_df = related_work_df[related_work_df['year'] != '-']
related_work_df['year'] = related_work_df['year'].astype(int)
related_work_df.sort_values(by=['year'], ascending=False, inplace=True)
related_work_df
```

```python
related_work_df =related_work_df.replace(np.nan, '-', regex=True)
```

```python
filter_attributes = ['Property']
cluster = related_work_df[related_work_df.isin(['K-Means', 'DBSCAN', 'DBSCAN, K-Means', 'AffinityPropagation', 'K-Prototypes']).any(1)]
cluster = cluster.drop(columns=filter_attributes)
cluster
```

```python
cluster.to_latex(index=False)
```

```python
dp = related_work_df[related_work_df.isin(['Differential privacy method']).any(1)]
dp = dp.drop(columns=filter_attributes)
dp
```
