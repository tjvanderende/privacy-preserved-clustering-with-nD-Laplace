```python
import numpy as np
import matplotlib.pyplot as plt

def laplace_pdf(r, theta, r0, theta0, epsilon):
    d = np.sqrt(r**2 + r0**2 - 2*r*r0*np.cos(theta - theta0))
    pdf = (epsilon**2 / (2 * np.pi)) * np.exp(-epsilon * d)
    return pdf

r = np.linspace(0, 5, 100)
theta = np.linspace(0, 2*np.pi, 100)
R, Theta = np.meshgrid(r, theta)
pdf = laplace_pdf(R, Theta, 1, np.pi/2, 1)

fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
ax.contour(Theta, R, pdf, cmap='viridis')
ax.set_title('Laplace PDF in Polar Coordinates')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some random points
n_points = 100
r_coords = np.random.uniform(0, 10, n_points)
theta_coords = np.random.uniform(0, 2*np.pi, n_points)

# Create a polar plot with radius as the x-axis and angle as the y-axis
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='polar')

# Draw the polar grid within the radius
theta_grid, r_grid = np.meshgrid(np.linspace(0, 2*np.pi, 100), np.linspace(0, np.max(r_coords), 10))
x_grid, y_grid = r_grid*np.cos(theta_grid), r_grid*np.sin(theta_grid)
ax.plot(theta_grid, r_grid, color='grey', linewidth=1, alpha=0.5)
ax.plot(x_grid, y_grid, color='grey', linewidth=1, alpha=0.5)

# Plot the random points with color based on distance from the origin
colors = r_coords / np.max(r_coords)
ax.scatter(theta_coords, r_coords, c=colors, cmap='viridis', alpha=0.8)

# Set the axis limits and labels
ax.set_ylim(0, np.max(r_coords))
ax.set_xticks(np.linspace(0, 2*np.pi, 12, endpoint=False))
ax.set_xticklabels(['0', '', '', '', '', '', '1', '', '', '', '', '2'])
ax.set_title('Scatterplot with Polar Grid')

# Show the plot
plt.show()
```
