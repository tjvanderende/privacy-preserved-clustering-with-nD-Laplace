```python
import numpy as np
import matplotlib.pyplot as plt

def laplace_pdf(r, theta, r0, theta0, epsilon):
    d = np.sqrt(r**2 + r0**2 - 2*r*r0*np.cos(theta - theta0))
    pdf = (epsilon**2 / (2 * np.pi)) * np.exp(-epsilon * d)
    return pdf

r = np.linspace(0, 5, 100)
theta = np.linspace(0, 2*np.pi, 100)
R, Theta = np.meshgrid(r, theta)
pdf = laplace_pdf(R, Theta, 1, np.pi/2, 1)

fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
ax.contour(Theta, R, pdf, cmap='viridis')
ax.set_title('Laplace PDF in Polar Coordinates')
plt.show()
```

```python
import numpy as np

# Define the number of grid cells in each dimension
num_cells_x = 10
num_cells_y = 10

# Define the latitude and longitude bounds of the geographic region
lat_min, lat_max = 40.0, 41.0
lon_min, lon_max = -74.0, -73.0

# Create a grid of cell boundaries
lat_edges = np.linspace(lat_min, lat_max, num_cells_y+1)
lon_edges = np.linspace(lon_min, lon_max, num_cells_x+1)

# Discretize the location data into grid cells
def discretize_locations(locations, lat_edges, lon_edges):
    cell_indices = np.zeros((len(locations), 2), dtype=int)
    for i, loc in enumerate(locations):
        cell_indices[i, 0] = np.searchsorted(lat_edges, loc[0], side='right') - 1
        cell_indices[i, 1] = np.searchsorted(lon_edges, loc[1], side='right') - 1
    return cell_indices

# Example usage
locations = [(40.6, -73.8), (40.1, -73.2), (40.8, -74.2)]
cell_indices = discretize_locations(locations, lat_edges, lon_edges)
print(cell_indices)
```
